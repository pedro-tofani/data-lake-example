# DEDUPLICATE DATE (MERGING)

```python
data_schema = spark.read.format('parquet').load("s3://s3-bucket-phtf-db/orders/public/customers/").schema
df = spark.read.format("parquet").schema(data_schema).load("s3://s3-bucket-phtf-db/orders/public/customers/")
df = df.withColumn("extracted_date", F.to_date(F.col("extracted_at")))
df = df.withColumn("created_date", F.to_date(F.col("created_at")))
df = df.withColumn("etl_timestamp", F.to_date(F.current_timestamp()))
df = df.withColumn("source_filename", F.input_file_name())

df.limit(1).write.format('delta').partitionBy('created_date').save("s3://s3-bucket-phtf-db-processed/orders/customers_merge")
```

```python
display(spark.read.format('delta').load("s3://s3-bucket-phtf-db-processed/orders/customers_merge"))
```

```python
from delta.tables import DeltaTable
from pyspark.sql.window import Window

from pyspark.sql import functions as F

deltaTable = DeltaTable.forPath(spark, path="s3://s3-bucket-phtf-db-processed/orders/customers_merge")

def _upsert(microbatch, batch_id):
  microbatch = microbatch.withColumn('row_rank', F.row_number().over(Window.partitionBy(F.col('customer_id')).orderBy(F.col('extracted_at').desc())))
  microbatch = microbatch.filter(F.col('row_rank') == 1)
  microbatch = microbatch.drop("row_rank")
  
  deltaTable.alias("target").merge(microbatch.alias("updates"), "target.customer_id = updates.customer_id").whenMatchedUpdateAll("updates.op='U'").whenNotMatchedInsertAll().execute()
  

data_schema = spark.read.format('parquet').load("s3://s3-bucket-phtf-db/orders/public/customers/").schema
df = spark.readStream.format("parquet").schema(data_schema).load("s3://s3-bucket-phtf-db/orders/public/customers/")
    
df = df.withColumn("extracted_date", F.to_date(F.col("extracted_at")))
df = df.withColumn("created_date", F.to_date(F.col("created_at")))
df = df.withColumn("etl_timestamp", F.to_date(F.current_timestamp()))
df = df.withColumn("source_filename", F.input_file_name())

df.writeStream.format('delta').option("checkpointLocation", "dbfs:/checkpoints/orders").outputMode('update').foreachBatch(_upsert).start()
```
